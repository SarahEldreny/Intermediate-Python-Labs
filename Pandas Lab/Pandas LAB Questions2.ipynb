# 5. Kindly handle the premier league top scorer's information as the following: 

i. Read the CSV file which contains the following information about top scorers: players names, rank, number of scored goals, and their nationalities

ii. Read the JSON file which contains the following information about top scorers: players names, height, appearances, wins, losses, goals per match ratio, assists, yellow cards, red cards 

iii. Merge the two datasets

iv. Visualize a scatter plot about the relation between number of matched played (Appearances) and the number of scored goals

```python
import pandas as pd
import matplotlib as plt
```

```python
# i.
df_csv = pd.read_csv('5 Check_top_scorers\Top_Scorers_dataset.csv', sep = ',')
df_csv
```

```python
#ii.
df_json = pd.read_json('5 Check_top_scorers\players_stats.json')
df_json
```

```python
result = pd.merge(df_csv, df_json, on='Name', how='outer')
result
```

```python
# iv. Visualize a scatter plot about the relation between 
#number of matched played (Appearances) and the number of scored goals
result.plot('Appearances', 'Goals_per_match',kind= 'scatter')
```
