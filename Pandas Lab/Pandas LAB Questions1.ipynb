# 2. Process the Titanic datasets separately as the following:

i. Read the CSV dataset considering “;” is the separator

ii. Read the JSON dataset considering “fields” only

iii. From the dataframe of CSV dataset, kindly remove “Age” & “Ticket” columns

iv. From the dataframe of CSV dataset, kindly remove “Fare” & “Cabin” columns

v. Combine the two dataframes into one

vi. Write the combined dataframe to CSV file

```python
# i.
df_csv = pd.read_csv('2 process_Titanic_datasets\\titanic-passengers.csv', sep = ';')
df_csv
```

```python
# ii.
df_json = pd.read_json('2 process_Titanic_datasets\\titanic-passengers.json')
df_json1 = pd.json_normalize(df_json['fields']) 
df_json1
```

```python
# iii.
df_csv1 = df_csv.drop(['Age', 'Ticket'], axis=1)
df_csv1
```

```python
# iv.
df_json1 = df_json1.drop(['fare', 'cabin'], axis=1)
df_json1
```

```python
# v.
df_json1 = df_json1['passengerid'].rename('PassengerId')
result = pd.merge(df_csv1, df_json1, on='PassengerId', how='outer')
result
```

```python
# vi.
result.to_csv("2 process_Titanic_datasets\\written_dataset.csv", index=False, sep=",")
```
