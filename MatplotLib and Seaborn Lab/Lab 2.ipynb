# Question #3:

Kindly use Seaborn heatmaps to visualize the
flights dataset.

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# Loading the data with pandas library and see info about it.
df = pd.read_csv("flights.csv")
df.head()
```

```python
#Loading the data with seaborn library

flights = sns.load_dataset("flights")


# The pivot function is used to create a new derived table from the given data frame object “df”. 
# The function takes three arguments; index, columns, and values. 
# The cell values of the new table are taken from column given as the values parameter,
# which in our case is the “passengers” column.

flights = flights.pivot("month", "year", "passengers")

#In the final step, we create the heatmap using the heatmap function from the seaborn python package. 
# The heatmap function takes the following arguments:
# data: 2D dataset that can be coerced into an ndarray. If a Pandas DataFrame is provided, the index/column information will be used to label the columns and rows
# annot: an array of same shape as data which is used to annotate the heatmap.
# fmt: string formatting code to use when adding annotations.

heatmap = sns.heatmap(flights, annot=True, fmt="d")
```

# Question #4:

Kindly use Seaborn relplot to visualize the mpg
dataset.

```python
#Loading the mpg data
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
# This plot shows the relationship between  variables in 
# the mpg dataset using a single call to the seaborn function relplot()

relplot = sns.relplot(data=mpg, x="horsepower", y="mpg", hue="origin", 
            size="weight", markers=True, kind="scatter")
```
