```python
import seaborn as sns

#sns.set_theme(style="ticks")
exercise = sns.load_dataset("exercise")
#exercise.head()
```

```python
g1 = sns.catplot(x="time", y="pulse", hue="kind", data=exercise)
```

```python
g2 = sns.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="violin")
```

```python
g3 = sns.catplot(x="time", y="pulse", hue="kind", col="diet", data=exercise, height=5, aspect=.8)
```
